```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/FazlyRabbiBD/Consultancy/main/Data-RDA-Training-Processed.csv")
df.head()
```

```python
pd.crosstab(df['Union'],df['Gender'],margins=True, margins_name="Total")
```

```python
plt.subplots(figsize=(6,6))
sns.set(font_scale=1.5)
sns.heatmap(pd.crosstab(index=df['Reason of unemployment'],
                    columns=df['Gender'],
                    normalize='all',margins=True, margins_name="Total"),
            cmap="YlGnBu", annot=True, cbar=False,fmt=".2%")
```

```python
pd.pivot_table(df, index=['Desiered Profession'], values=['Age'], aggfunc=['median'])
```

```python
pd.crosstab(df['M_Status'],df['Gender'],margins=True, margins_name="Total")
```

```python
plt.subplots(figsize=(6,6))
sns.set(font_scale=1.5)
sns.heatmap(pd.crosstab(index=df['Education'],
                    columns=df['Gender'],
                    normalize='all',margins=True, margins_name="Total"),
            cmap="YlGnBu", annot=True, cbar=False,fmt=".2%")
```
